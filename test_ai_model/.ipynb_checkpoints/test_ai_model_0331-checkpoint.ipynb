```python
import sys
import os

# 'models' 디렉토리를 sys.path에 추가
sys.path.append(os.path.abspath('../models'))
```

```python
import numpy as np
import torch
import torch.nn as nn

# 1. 먼저 전체 데이터를 하나의 리스트로 읽기
users, movies, ratings = [], [], []
user_len, movie_len = 0, 0

with open("../data/ratings.dat", "r") as f:
    for line in f:
        u, m, r, t = line.split("::")
        user_len = max(user_len, int(u))
        movie_len = max(movie_len, int(m))
        users.append(int(u))
        movies.append(int(m))
        ratings.append(float(r))

indices = np.arange(len(ratings))
np.random.shuffle(indices)

users = np.array(users)[indices]
movies = np.array(movies)[indices]
ratings = np.array(ratings)[indices]

device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")


users = torch.tensor(users, dtype=torch.long, device="mps")
movies = torch.tensor(movies, dtype=torch.long, device = "mps")
ratings = torch.tensor(ratings, dtype=torch.float, device = "mps")

# 3. 데이터 분할 (예: 60% 학습, 20% 검증, 20% 테스트)
n_samples = len(ratings)
train_size = int(0.6 * n_samples)
valid_size = int(0.2 * n_samples)

# 학습 데이터
users_train = users[:train_size]
movies_train = movies[:train_size]
ratings_train = ratings[:train_size]


# 검증 데이터
users_valid = users[train_size:train_size+valid_size]
movies_valid = movies[train_size:train_size+valid_size]
ratings_valid = ratings[train_size:train_size+valid_size]


# 테스트 데이터
users_test = users[train_size+valid_size:]
movies_test = movies[train_size+valid_size:]
ratings_test = ratings[train_size+valid_size:]




```

```python
from torch.utils.data import DataLoader, TensorDataset

train_loader = DataLoader(TensorDataset(users_train, movies_train, ratings_train), batch_size=64, shuffle=True)
valid_loader = DataLoader(TensorDataset(users_valid, movies_valid, ratings_valid), batch_size=64, shuffle=False)
test_loader = DataLoader(TensorDataset(users_test, movies_test, ratings_test), batch_size=64, shuffle=False)
```

```python
class DL_MF(nn.Module):
    def __init__(self, user_len, movie_len,rank):
        super(DL_MF,self).__init__()
        self.user_len = user_len
        self.movie_len = movie_len
        self.rank = rank
        
        self.users_embs = nn.Embedding(user_len+1, rank)
        self.items_embs = nn.Embedding(movie_len+1, rank)
        
        self.fc1 = nn.Linear(rank*2, rank)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(rank, 5)
        self.fc3 = nn.Linear(5, 1)
        
    def forward(self, user_idx, movie_idx):
        user_emb = self.users_embs(user_idx)
        item_emb = self.items_embs(movie_idx)
        user_item_concat_emb = torch.concat([user_emb, item_emb], dim=1)
        x = self.fc1(user_item_concat_emb)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.relu(x)
        x = self.fc3(x)  # 최종 출력 (batch_size, 1)
        
        return x.squeeze()  # (batch_size,)
```

```python
import torch.nn as nn
import torch.optim as optim
import torch

class MF(nn.Module):
    def __init__(self, user_len, movie_len, ratings, rank):
        super(MF, self).__init__()
        self.user_len = user_len
        self.movie_len = movie_len
        self.rank = rank
        self.alpha = ratings.float().mean()
        self.user_bias = nn.Embedding(user_len+1, 1)
        self.item_bias = nn.Embedding(movie_len+1, 1)
        self.user_reps = nn.Embedding(user_len+1, rank)
        self.item_reps = nn.Embedding(movie_len+1, rank)
        
        nn.init.xavier_uniform_(self.user_reps.weight)
        nn.init.xavier_uniform_(self.item_reps.weight)
        nn.init.zeros_(self.user_bias.weight)
        nn.init.zeros_(self.item_bias.weight)

    def forward(self, user,item):
        user_bias = self.user_bias(user).squeeze()
        item_bias = self.item_bias(item).squeeze()
        user_emb = self.user_reps(user)
        item_emb = self.item_reps(item)
        return self.alpha + user_bias + item_bias + torch.sum(user_emb*item_emb,dim=1)
    

    
    
```

```python
import time
import numpy as np
import torch
import torch.optim as optim
import torch.nn as nn
import matplotlib.pyplot as plt

# Hyperparameters
epochs = 10
dl_mf_rank = 10
mf_rank = 10


device = torch.device("mps" if torch.mps.is_available() else "cpu")

# 모델 1: DL_MF
model_1 = DL_MF(user_len, movie_len, dl_mf_rank).to(device)
optimizer_1 = optim.Adam(model_1.parameters(), lr=0.001)
criterion = nn.MSELoss()

# 모델 2: MF
model_2 = MF(user_len, movie_len, ratings_train, mf_rank).to(device)  # MF 모델은 예시로 설정한 모델입니다
optimizer_2 = optim.Adam(model_2.parameters(), lr=0.001)

# 기록할 값들
epoch_train_losses_dl_mf, epoch_val_losses_dl_mf = [], []
epoch_train_losses_mf, epoch_val_losses_mf = [], []

# 테스트 손실을 마지막에만 저장
test_loss_dl_mf, test_loss_mf = None, None

for epoch in range(epochs):
    start_time = time.time()  # 각 에폭의 시작 시간 기록
    
    # Training for DL_MF (모델 1)
    model_1.train()
    train_losses_dl_mf = []
    for user, movie, rating in train_loader:
        user, movie, rating = user.to(device), movie.to(device), rating.to(device)
        
        optimizer_1.zero_grad()
        pred_1 = model_1(user, movie)
        loss_1 = criterion(pred_1, rating)
        train_losses_dl_mf.append(loss_1.item())
        
        loss_1.backward()
        optimizer_1.step()

    # Validation for DL_MF
    model_1.eval()
    valid_losses_dl_mf = []
    with torch.no_grad():
        for user, movie, rating in valid_loader:
            user, movie, rating = user.to(device), movie.to(device), rating.to(device)
            pred_1 = model_1(user, movie)
            loss_1 = criterion(pred_1, rating)
            valid_losses_dl_mf.append(loss_1.item())

    # Training for MF (모델 2)
    model_2.train()
    train_losses_mf = []
    for user, movie, rating in train_loader:
        user, movie, rating = user.to(device), movie.to(device), rating.to(device)
        
        optimizer_2.zero_grad()
        pred_2 = model_2(user, movie)
        loss_2 = criterion(pred_2, rating)
        train_losses_mf.append(loss_2.item())
        
        loss_2.backward()
        optimizer_2.step()

    # Validation for MF
    model_2.eval()
    valid_losses_mf = []
    with torch.no_grad():
        for user, movie, rating in valid_loader:
            user, movie, rating = user.to(device), movie.to(device), rating.to(device)
            pred_2 = model_2(user, movie)
            loss_2 = criterion(pred_2, rating)
            valid_losses_mf.append(loss_2.item())

    # 에폭 결과 출력
    epoch_train_loss_dl_mf = np.mean(train_losses_dl_mf)
    epoch_val_loss_dl_mf = np.mean(valid_losses_dl_mf)

    epoch_train_loss_mf = np.mean(train_losses_mf)
    epoch_val_loss_mf = np.mean(valid_losses_mf)

    # 결과를 리스트에 추가
    epoch_train_losses_dl_mf.append(epoch_train_loss_dl_mf)
    epoch_val_losses_dl_mf.append(epoch_val_loss_dl_mf)

    epoch_train_losses_mf.append(epoch_train_loss_mf)
    epoch_val_losses_mf.append(epoch_val_loss_mf)

    epoch_time = time.time() - start_time  # 에폭이 끝난 후 시간 계산

    print(f"Epoch {epoch+1}/{epochs}, Time: {epoch_time:.2f}s")
    print(f"Model 1 (DL_MF) - Train Loss: {epoch_train_loss_dl_mf:.4f}, Val Loss: {epoch_val_loss_dl_mf:.4f}")
    print(f"Model 2 (MF) - Train Loss: {epoch_train_loss_mf:.4f}, Val Loss: {epoch_val_loss_mf:.4f}")

# 테스트셋에서 최종 손실 계산 (한 번만)
model_1.eval()
test_losses_dl_mf = []
with torch.no_grad():
    for user, movie, rating in test_loader:
        user, movie, rating = user.to(device), movie.to(device), rating.to(device)
        pred_1 = model_1(user, movie)
        loss_1 = criterion(pred_1, rating)
        test_losses_dl_mf.append(loss_1.item())

model_2.eval()
test_losses_mf = []
with torch.no_grad():
    for user, movie, rating in test_loader:
        user, movie, rating = user.to(device), movie.to(device), rating.to(device)
        pred_2 = model_2(user, movie)
        loss_2 = criterion(pred_2, rating)
        test_losses_mf.append(loss_2.item())

test_loss_dl_mf = np.mean(test_losses_dl_mf)
test_loss_mf = np.mean(test_losses_mf)

# 최종 테스트 결과 출력
print("\nModel 1 (DL_MF) Test Loss:", test_loss_dl_mf)
print("Model 2 (MF) Test Loss:", test_loss_mf)

# 훈련과 검증 손실 그래프 그리기
plt.figure(figsize=(12, 6))

# DL_MF 모델
plt.subplot(1, 2, 1)
plt.plot(range(epochs), epoch_train_losses_dl_mf, label='Train Loss')
plt.plot(range(epochs), epoch_val_losses_dl_mf, label='Val Loss')
plt.title('DL_MF Model')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# MF 모델
plt.subplot(1, 2, 2)
plt.plot(range(epochs), epoch_train_losses_mf, label='Train Loss')
plt.plot(range(epochs), epoch_val_losses_mf, label='Val Loss')
plt.title('MF Model')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```
